```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = {'x': np.random.rand(100), 'y': np.random.rand(100)}
idx = 50
```

```python
data = {'x': np.random.rand(1000), 'y': np.random.rand(1000)}
idx = 50

points = np.column_stack((data['x'], data['y']))
distances = np.linalg.norm(points - points[idx], axis=1)

# Get indices of the 10 nearest neighbors (excluding itself)
nearest_indices = np.argsort(distances)[1:101]

# Plot the points and highlight the neighbors
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot()
sc = ax.scatter(data['x'], data['y'], s=15, alpha=1)
ax.scatter(data['x'][idx], data['y'][idx], color='red', s=30, label='Center Point')
ax.scatter(data['x'][nearest_indices[:5]], data['y'][nearest_indices[:5]], color='green', s=30, label='Nearest Neighbors')
ax.scatter(data['x'][nearest_indices[95:]], data['y'][nearest_indices[95:]], color='yellow', s=30, label='Cluster Border')

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.legend()
plt.show()
```
